```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/NHANES.csv')
df = df[['Gender', 'Age', 'Height']]
df = df.dropna()
```

```python
adult = df['Age'] >= 20


male = df['Gender'] == 'male'
female = df['Gender'] == 'female'


male_df = df[adult & male]
female_df = df[adult & female]
```

```python
male_height = male_df['Height']
female_height = female_df['Height']
```

```python
fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot([male_height, female_height])
plt.title('Height distribution of the U.S.', fontsize=20)
axs.set_xticklabels(['Male Adult','Female Adult'])
axs.set_ylabel('Height [cm]', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

```python

height = male_df['Height']

maximum = np.max(height)
minimum = np.min(height)
median = np.median(height)

q1 = np.percentile(height, 25, interpolation='midpoint')
q3 = np.percentile(height, 75, interpolation='midpoint')

print("Max male adult height:", maximum)
print("Min male adult height:", minimum)
print("The median is:", median)
print("The interquartile range is:", q3-q1)
```
